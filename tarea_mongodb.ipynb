```python
import pymongo
import json
import pandas as pd
```

```python
myclient = pymongo.MongoClient("mongodb://localhost:27017/")

db = myclient["tarea_pymongo"]

cryptopunks = db["cryptopunks"]
```

```python
with open("cryptopunks.json") as file:
    file_data = json.load(file)

cryptopunks.insert_many(file_data)
```

```python
find_all = db.cryptopunks.find()
df_find_all = pd.DataFrame(find_all)
print(df_find_all)
```

```python
count_total = db.cryptopunks.find().sort("punk_id",-1).limit(10)
df_count_total = pd.DataFrame(count_total)
print(df_count_total)
```
